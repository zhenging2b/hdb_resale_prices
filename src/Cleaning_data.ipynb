## Found 5 Jln Batu wrongly mapped to bus depot

```python
import requests
response = requests.get(
    ONEMAP_SEARCH_URL, 
    params = {"searchVal":"5 Jln Batu", "returnGeom": "Y", "getAddrDetails": "Y", "pageNum": 1,},
    headers= {"Authorization": token},
    timeout=60
)   
response
response.json()
```

```python
geo_loc.loc[geo_loc["address"] == "5 JLN BATU"]
```

```python
dic = response.json()["results"][1]
```

```python
dic
```

```python
update_lst = {
    "onemap_searchval": dic.get("SEARCHVAL"),
    "onemap_blk_no": dic.get("BLK_NO"),
    "onemap_road_name": dic.get("ROAD_NAME"),
    "onemap_building": dic.get("BUILDING"),
    "onemap_address": dic.get("ADDRESS"),
    "onemap_postal": dic.get("POSTAL"),
    "onemap_x": float(dic["X"]) if dic.get("X") else None,
    "onemap_y": float(dic["Y"]) if dic.get("Y") else None,
    "latitude": float(dic["LATITUDE"]),
    "longitude": float(dic["LONGITUDE"]),
}
```

```python
geo_loc.loc[geo_loc["address"] == "5 JLN BATU", list(update_lst.keys())] = list(update_lst.values())
```

```python
geo_loc.to_csv(DATA_DIR / "geocoded_addresses.csv")
```

## Merge to planning area

Merge to plannign area for different regions

```python
df_planning_area= pd.read_csv(DATA_DIR / "SG_planning_areas.csv")
```

```python
df_planning_area["town_cleaned"] = df_planning_area.Town.str.upper()
df_planning_area.loc[df_planning_area["Town"] == "Jurong East • Jurong Timur", "town_cleaned"] = "JURONG EAST"
df_planning_area.loc[df_planning_area["Town"] == "Jurong West • Jurong Barat", "town_cleaned"] = "JURONG WEST"
df_planning_area.loc[df_planning_area["Town"] == "Kallang", "town_cleaned"] = "KALLANG/WHAMPOA"
```

```python
import numpy as np
```

```python
df_planning_area.loc[len(df_planning_area)] = ["Central Area", "Central", 17.84, np.nan, np.nan, np.nan, "CENTRAL AREA"]
```

```python
df_planning_area
```

```python
df_geo = pd.read_csv(DATA_DIR/ "geocoded_addresses.csv")
```

```python
df_merge_geo = df_total.merge(df_geo, how="left", on="address")
final_df = df_merge_geo.merge(df_planning_area, how='left', left_on="town", right_on="town_cleaned")
```

```python
final_df.to_csv(DATA_DIR / "geocode_final.csv",index=False)
```
